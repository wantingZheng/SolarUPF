```python
import numpy as np
import pandas as pd
from calculate_metric import calculate_cgre,get_metric_block,plot_performance_metrics_with_top_labels_centered
```

## 30days_test

```python
metric_name='NRMSE'
file_path="testday_30_interval_result1.xlsx"
file_path_ext="external_testday_30_interval_result1.xlsx"
long_df=get_metric_block(file_path, metric_name)
long_df_ext=get_metric_block(file_path_ext, metric_name)
capacities = pd.Series(
    [37289.2517,17792.2276,801.0904,200.0827,72.2296,5197.996,469.6753,10706.4215,28.4859,5170.437,332.0071,
246.4706,39218.28513,614153.6403,11900.635,23586.4859,17736.6813,19787.00779,37092.6257,19338.2813,48184.2682,48321.73,
109360.3,114688,29410,212.2530646,645.8106825,126.9360217,842.2084351,44.99863052,23.64752532,107.8802235,37.8619194,
979.9,979.975,1003.55],
    index=long_df.index
)

result, details = calculate_cgre(
    data=long_df,
    capacities=capacities,
    theta=(0.4, 0.3, 0.3),   #  θ1, θ2, θ3
    beta=0.9)
result_ext, details_ext = calculate_cgre(
    data=long_df_ext,
    capacities=capacities,
    theta=(0.4, 0.3, 0.3),   #  θ1, θ2, θ3
    beta=0.9)

print(result)   
print(result_ext)                         # A_opt, A_cap, A_risk, G
result_test_ext=(result+result_ext)/2
result_test_ext_NRMSE1 = result_test_ext.round(4)
result_test_ext_NRMSE = result_test_ext.sort_values("G", ascending=True).round(4)
print(result_test_ext_NRMSE)  
```
